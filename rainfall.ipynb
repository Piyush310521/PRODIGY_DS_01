```python
yeanrain = swise.groupby('Year')['N Heavy Rainfall Events'].mean()
```

```python
yeanrain
```

```python
df = yeanrain.reset_index()
```

```python
df
```

```python

plt.bar(df['Year'], df['N Heavy Rainfall Events'])

plt.xlabel('Year')
plt.ylabel('Number of Heavy Rainfall Events')
plt.title('Heavy Rainfall Events per Year')
plt.show()
```

The bar graph illustrates the number of Heavy rainfall events recorded each month from 2017 to 2022. It is evident that 2021 experienced a higher frequency of rainfall events compared to the previous year. This suggests a notable increase in precipitation occurrences during the specified time period. and the lowest frequency of Rainfall has happened in the year 2018. Point to remember this analysis is only on the rainfall happened from year 2017-2022.
