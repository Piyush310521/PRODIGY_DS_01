```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
swise = pd.read_csv('YEARWISERAINFALLHVE.csv')
```

```python
swise
```

```python
newc = {'Number of Heavy Rainfall Events': 'N Heavy Rainfall Events', 'Number of Very Heavy Rainfall Events': 'N Very Heavy Rainfall Events', 'Number of Extremely Heavy Rainfall Events': 'N Extremly Heavy Rainfall Events'} 
swise.rename(columns = newc, inplace = True )
```

```python
def heavyrain():
    _2017 = swise.loc[0:7, 'N Heavy Rainfall Events'].sum()
    print("heavy rainfall 2017",_2017)
    _2018 = swise.loc[8:15, 'N Heavy Rainfall Events'].sum()
    print("heavy rainfall 2018",_2018)
    _2019 = swise.loc[16:23, 'N Heavy Rainfall Events'].sum()
    print("heavy rainfall 2019",_2019)
    _2020 = swise.loc[24:31, 'N Heavy Rainfall Events'].sum()
    print("heavy rainfall 2020",_2020)
```

```python
heavyrain()
```

Since we have the highest number of heavy rainfall events in the YEAR 2019, we can say that the frequency of heavy rainfall happpened in year 2019.

```python
yeanrain = swise.groupby('Year')['N Heavy Rainfall Events'].mean()
```

```python
yeanrain
```

```python
df = yeanrain.reset_index()
```

```python
df
```

```python
plt.bar(df['Year'], df['N Heavy Rainfall Events'])

plt.xlabel('Year')
plt.ylabel('Number of Heavy Rainfall Events')
plt.title('Heavy Rainfall Events per Year')
plt.savefig('Heavy rainfall_events_per_year.png')
plt.show()
```

The bar graph illustrates the number of Heavy rainfall events recorded each month from 2017 to 2022. It is evident that 2021 experienced a higher frequency of rainfall events compared to the previous year. This suggests a notable increase in precipitation occurrences during the specified time period. and the lowest frequency of Rainfall has happened in the year 2018. Point to remember this analysis is only on the rainfall happened from year 2017-2022.
